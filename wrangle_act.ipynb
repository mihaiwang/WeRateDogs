```python
#连接twitter api
consumer_key = '(hidden)'
consumer_secret = '(hidden)'
access_token = '(hidden)'
access_secret = '(hidden)'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
```

```python
#定义读取json函数
def get_json(id_list,error_list):
    for _id in id_list:
        try:
            tweet=api.get_status(id=_id,tweet_mode='extended')._json
        except:
            error_list.append(_id)
            print(_id)
            print(sys.exc_info()[1])
            continue
        with open("tweet_json.txt","a") as outfile:
            json.dump(tweet,outfile)
            outfile.writelines('\n')
    print('error_list:',error_list,'\nerror_number:',len(error_list))
```

```python
#读取id列对应的json文件，存入id_list
id_list=[]
for line in pd.read_csv('twitter-archive-enhanced.csv').tweet_id:
    id_list.append(line)

not_found_list=[]
get_json(id_list,not_found_list)
```

```python
#再次传入not_found_list，排除随机出现的错误
not_found_again_list=[]
get_json(not_found_list,not_found_again_list)
```

```python
#再次传入not_found_again_list，排除随机出现的错误
not_found_thrice_list=[]
get_json(not_found_again_list,not_found_thrice_list)
```

```python
#从expanded_urls列中提取id
#检查确认：提取的derived_id与text,name一致，可作为原id的替代来源
#传入amend_dict字典

csv_fh=pd.read_csv('twitter-archive-enhanced.csv')
csv_fh['derived_id']=csv_fh.expanded_urls.str.extract('status/(\d{18})',expand=False)
csv_fh.to_csv('twitter-archive-enhanced-new.csv')

amend_dict={}
for not_found in not_found_thrice_list:
    for i in range(len(csv_fh.tweet_id)):
        if csv_fh.tweet_id[i] == not_found:
            print(csv_fh.expanded_urls.str.extract('status/(\d{18})',expand=False)[i],not_found)
            amend_dict[int(csv_fh.expanded_urls.str.extract('status/(\d{18})',expand=False)[i])]=not_found
print(amend_dict)
```

#### `twitter-archive-enhanced`中'doggo','floofer','pupper','puppo'应是‘stage’列的变量

##### 定义
从text中提取'doggo','floofer','pupper','puppo'关键字存入stage列

##### 代码

```python
def getstage(x,y):
    if x==-1:
        return ' '
    else:
        return y

df_tae['stage']= \
  df_tae.text.apply(lambda x:x.lower()).str.find('doggo').apply(lambda x:getstage(x,'doggo')) \
+ ' ' + df_tae.text.apply(lambda x:x.lower()).str.find('floofer').apply(lambda x:getstage(x,'floofer'))\
+ ' ' + df_tae.text.apply(lambda x:x.lower()).str.find('pupper').apply(lambda x:getstage(x,'pupper'))\
+ ' ' + df_tae.text.apply(lambda x:x.lower()).str.find('puppo').apply(lambda x:getstage(x,'puppo'))
df_tae[['text','stage','doggo','floofer','pupper','puppo']]
```

```python
df_tae.stage=df_tae.stage.str.strip().replace(re.compile('(.*\s+.*)'),'multiple')
df_tae.stage=df_tae.stage.apply(lambda x: np.NaN if str(x)=='' else x)
```

```python
#检查原4列中非None值的行是否在stage中也有值
s1=df_tae.doggo != 'None'
s2=df_tae.floofer != 'None'
s3=df_tae.pupper != 'None'
s4=df_tae.puppo != 'None'
df_tae[s1|s2|s3|s4].stage.isnull().value_counts()
```

```python
df_tae=df_tae.drop(['doggo','floofer','pupper','puppo'],axis=1)
```

#### `twitter-archive-enhanced`  'rating_denominator'有非10的项

##### 定义
从text中截取10的倍数的分母及其分子，放入rating_denominator和rating_numerator

##### 代码

```python
pd.set_option('max_colwidth', 200)
```

```python
df_tae[df_tae.rating_denominator!=10].text
```

```python
df_tae['rating_numerator']=df_tae.text.str.extract('((\d+\.)?\d+)/([1-9]+0)',expand=True)[0]
df_tae['rating_denominator']=df_tae.text.str.extract('(\d+)/([1-9]+0)',expand=True)[1]

df_tae.loc[df_tae.rating_numerator.notnull(),'rating_numerator']=\
df_tae.loc[df_tae.rating_numerator.notnull(),'rating_numerator'].astype('float')
df_tae.loc[df_tae.rating_numerator.notnull(),'rating_denominator']=\
df_tae.loc[df_tae.rating_numerator.notnull(),'rating_denominator'].astype('float')
```
